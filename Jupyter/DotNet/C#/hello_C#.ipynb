```
int[] arr = { 1, 3, 5, 7, 9, 0, 2, 4, 6, 8 };
var query = from n in arr
            group n by n % 3;

foreach (var i in query)
{
    foreach (var j in i)
    {
        Console.WriteLine($"{i.Key}\t{j}");
    }
}
```

```
string[] langs = { "C#", "Go", "Python", "C", "C++", "Java", "Delphi", "VB.net", "Matlab", "R", "Ruby", "Perl" };
var query = from lang in langs
            group lang by lang.Length into lenGroups
            orderby lenGroups.Key
            select lenGroups;

foreach (var group in query)
{
    foreach (var item in group)
    {
        Console.WriteLine($"{group.Key}:\t{group.Count()}\t{item}\t{group.Aggregate((s1, s2) => s1 + "," + s2)}");
    }
}
//or
foreach (var group in query)
{
    foreach (var item in group)
    {
        Console.WriteLine($"{group.Key}:\t{group.Count()}\t{item}\t{string.Join(",", group.Select(s => s))}");
    }
}
```
