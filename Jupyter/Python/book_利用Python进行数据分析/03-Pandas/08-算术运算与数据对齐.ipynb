```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
s1 + s2  # 算术运算，按照索引对齐数据（结果索引是两边索引的并集），类似于t-sql中的full outer join
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list('bcd'),
                   index=['Ohio', 'Texas', 'Colorado'])
"""
            b    c    d
Ohio      0.0  1.0  2.0
Texas     3.0  4.0  5.0
Colorado  6.0  7.0  8.0
"""
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list('bde'),
                   index=['Utah', 'Ohio', 'Texas', 'Oregon'])
"""
          b     d     e
Utah    0.0   1.0   2.0
Ohio    3.0   4.0   5.0
Texas   6.0   7.0   8.0
Oregon  9.0  10.0  11.0
"""
df1 + df2  # DataFrame会同时对齐行和列
```

```python
# 将两个行或列完全不同的DataFrame相加，结果为空
df1 = pd.DataFrame({'A': [1, 2]})
df2 = pd.DataFrame({'B': [3, 4]})
df1 + df2
```

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)), columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)), columns=list('abcde'))
df2.loc[1, 'b'] = np.nan
df1 + df2  # 对齐会导致结果中出现一些NaN
df1.add(df2, fill_value=0)  # 缺失位置的值使用0进行运算，类似于t-sql中的isnull(x, 0)
```

```python
# 常用算术方法
# 方法                   说明
# add, radd            用于加法（+）的方法
# sub, rsub            用于减法（-）的方法
# div, rdiv            用于除法（/）的方法
# floordiv, rfloordiv  用于底除（〃）的方法
# mul, rmul            用于乘法（*）的方法
# pow, rpow            用于指数（**）的方法
```

```python
# 上表中的r方法是翻转方法
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)), columns=list('abcd'))
df1.rdiv(1)  # <==>  1 / df1
```

```python
# DataFrame与Series之间的运算与NumPy的广播机制类似，默认使用Series的索引与DataFrame的列索引匹配，然后广播到各行
frame = pd.DataFrame(np.arange(12.).reshape((4, 3)),
                     columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
series = frame.iloc[0]
frame - series
```

```python
frame = pd.DataFrame(np.arange(12.).reshape((4, 3)),
                     columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
series = pd.Series(range(3), index=['b', 'e', 'f'])
frame + series
```

```python
frame = pd.DataFrame(np.arange(12.).reshape((4, 3)),
                     columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
series = frame['d']
frame.sub(series, axis='index')  # <==> axis=0，使用算术方法可以实现按行匹配，在列上广播
```
