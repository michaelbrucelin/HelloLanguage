```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('../data/examples/tips.csv')
tips['tip_pct'] = tips['tip']/tips['total_bill']  # 增加一个消费比例列
tips[:6]
```

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]

top(tips, n=6)  # 小费百分比最高的6组
```

```python
tips.groupby('smoker').apply(top)  # 获取了每个分组的小费百分比最高的5组
```

```python
tips.groupby('smoker', group_keys=False).apply(top)  # 禁用分组键
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')  # 给函数传递参数
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
result
```

```python
result.unstack('smoker')
```

```python
# 在GroupBy中，当你调用诸如describe之类的方法时，实际上只是应用了下面两条代码的快捷方式而已：
f = lambda x: x.describe()
tips.groupby('smoker')['tip_pct'].apply(f)
```
