```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime
```

```python
close_px_all = pd.read_csv('../data/examples/stock_px_2.csv', parse_dates=True, index_col=0)
close_px = close_px_all[['AAPL', 'MSFT', 'XOM']]
close_px = close_px.resample('B').ffill()
close_px
```

```python
close_px.AAPL.plot()
close_px.AAPL.rolling(250).mean().plot()  # 前250个值的平均值，最前面249项由于对应前面不足250个值，所以没有结果
# rolling运算符与resample和groupby很像，类似于t-sql中的窗口函数
```

```python
appl_std250 = close_px.AAPL.rolling(250, min_periods=10).std()
display(appl_std250[5:12])
appl_std250.plot()
```

```python
expanding_mean = appl_std250.expanding().mean()
expanding_mean.plot()
```

```python
close_px.rolling(60).mean().plot(logy=True)  # 窗口函数会应用到DataFrame的每一列上
```

```python
close_px.rolling('20D').mean().plot()  # rolling()还可以接收固定大小的时间偏移
```

### 指数加权函数

```python
aapl_px = close_px.AAPL['2006':'2007']
ma60 = aapl_px.rolling(30, min_periods=20).mean()
ewma60 = aapl_px.ewm(span=30).mean()

# 简单移动平均与加权平均的对比
ma60.plot(style='k--', label='Simple MA')
ewma60.plot(style='k--', label='EW MA')
plt.legend()
```

### 二元移动窗口函数

```python
spx_px = close_px_all['SPX']
spx_rets = spx_px.pct_change()
returns = close_px.pct_change()

corr = returns.AAPL.rolling(125, min_periods=100).corr(spx_rets)
corr.plot()
```

```python
corr = returns.rolling(125, min_periods=100).corr(spx_rets)
corr.plot()
```

### 用户自定义的移动窗口函数

rolling_apply函数使你能够在移动窗口上应用自己设计的数组函数。唯一要求的就是：该函数要能从数组的各个片段中产生单个值（即缩聚）。

```python
from scipy.stats import percentileofscore

score_at_2percent = lambda x: percentileofscore(x, 0.02)
result = returns.AAPL.rolling(250).apply(score_at_2percent)
result.plot()
```
