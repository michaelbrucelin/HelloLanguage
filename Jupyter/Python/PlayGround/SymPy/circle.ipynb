```python
x, y = symbols('x,y')
func1 = sympy.Eq(y**2 + x**2, 1**2)
func2 = sympy.Eq(y**2 + x**2, 2**2)
func3 = sympy.Eq(y**2 + x**2, 3**2)
func4 = sympy.Eq(y**2 + x**2, 4**2)
func5 = sympy.Eq(y**2 + x**2, 5**2)
func6 = sympy.Eq(y**2 + x**2, 6**2)
func7 = sympy.Eq(y**2 + x**2, 7**2)
func8 = sympy.Eq(y**2 + x**2, 8**2)
func9 = sympy.Eq(y**2 + x**2, 9**2)
func10 = sympy.Eq(y**2 + x**2, 10**2)

plot1 = sympy.plot_implicit(func1, xlim=(-1, 1), ylim=(-1, 1), show=False)
plot2 = sympy.plot_implicit(func2, xlim=(-2, 2), ylim=(-2, 2), show=False)
plot3 = sympy.plot_implicit(func3, xlim=(-3, 3), ylim=(-3, 3), show=False)
plot4 = sympy.plot_implicit(func4, xlim=(-4, 4), ylim=(-4, 4), show=False)
plot5 = sympy.plot_implicit(func5, xlim=(-5, 5), ylim=(-5, 5), show=False)
plot6 = sympy.plot_implicit(func6, xlim=(-6, 6), ylim=(-6, 6), show=False)
plot7 = sympy.plot_implicit(func7, xlim=(-7, 7), ylim=(-7, 7), show=False)
plot8 = sympy.plot_implicit(func8, xlim=(-8, 8), ylim=(-8, 8), show=False)
plot9 = sympy.plot_implicit(func9, xlim=(-9, 9), ylim=(-9, 9), show=False)
plot10 = sympy.plot_implicit(func10, xlim=(-10, 10), ylim=(-10, 10), show=False)

plot = sympy.plot(xlim=(-11, 11), ylim=(-11, 11), aspect_ratio=(1., 1.), show=False)
plot.extend([plot1[0], plot2[0], plot3[0], plot4[0], plot5[0], 
             plot6[0], plot7[0], plot8[0], plot9[0], plot10[0]])

plot.show()
```
