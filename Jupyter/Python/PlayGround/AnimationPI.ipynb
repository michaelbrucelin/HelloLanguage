```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# https://www.bilibili.com/video/BV1TX4y1c7sa，无意间发现这个小视频，觉得挺有意思，就实现一下玩玩
```

```python
# 线速度相等版
# 无论怎样调整，都是不对的，想想就想明白了，调整也只是影响速率而已
from matplotlib import animation
import matplotlib.patches as mpathes
from IPython.display import HTML
plt.rcParams['animation.embed_limit'] = 2**10

N = 19
UNITPI = np.pi*2/(N-1)  # 最内圈角速度

fig, ax = plt.subplots()
ax.set_xlim((-N, N))
ax.set_ylim((-N, N))
ax.set_aspect('equal')

for i in range(1, N):
    ax.add_patch(mpathes.Circle((0,0) , i, fill=False, alpha=0.1))  # 添加同心圆

linepi, = plt.plot(np.cos(-UNITPI*0/np.arange(1, N))*np.arange(1, N),
                   np.sin(-UNITPI*0/np.arange(1, N))*np.arange(1, N),
                   'r-',
                   marker='o', markersize=4)

def update(i):
    linepi.set_data(np.cos(-UNITPI*i/np.arange(1, N))*np.arange(1, N),
                    np.sin(-UNITPI*i/np.arange(1, N))*np.arange(1, N))
    return linepi,

anim = animation.FuncAnimation(fig, update, frames=201, interval=50, blit=True, repeat=True)

plt.show()
# anim.save('pi.gif')
# HTML(anim.to_html5_video())
# HTML(anim.to_jshtml())  # 产生的文件太大，测试时可以打开
```

```python
# 角速度倍增，假定最外圈角速度为θ，则倒数第二圈为2θ，倒数第三圈为3θ... ...依此类推
"""
B站用户的评论：
由内到外小球公转周期递增，设为T1，T2，T3...Tn，从同一侧共线处开始运行，
当运行时间t为T1，T2，T3...Tn的最小公倍数的整数倍时小球再次在同一侧共线，
最小公倍数亦为小球同一侧共线的最小正周期，就是在一系列半径等差的同心圆上，各个点在同一条直线上以相同的线速度起步得到的
"""
from matplotlib import animation
import matplotlib.patches as mpathes
from IPython.display import HTML
plt.rcParams['animation.embed_limit'] = 2**10

N = 19
UNITPI = np.pi/100  # 最外圈角速度

fig, ax = plt.subplots()
ax.set_xlim((-N, N))
ax.set_ylim((-N, N))
ax.set_aspect('equal')

for i in range(1, N):
    ax.add_patch(mpathes.Circle((0,0) , i, fill=False, alpha=0.1))  # 添加同心圆

linepi, = plt.plot(np.cos(-UNITPI*0*np.arange(N-1, 0, -1))*np.arange(1, N),
                   np.sin(-UNITPI*0*np.arange(N-1, 0, -1))*np.arange(1, N),
                   'r-',
                   marker='o', markersize=4)

def update(i):
    linepi.set_data(np.cos(-UNITPI*i*np.arange(N-1, 0, -1))*np.arange(1, N),
                    np.sin(-UNITPI*i*np.arange(N-1, 0, -1))*np.arange(1, N))
    return linepi,

anim = animation.FuncAnimation(fig, update, frames=201, interval=50, blit=True, repeat=True)

plt.show()
# anim.save('pi.gif')
# HTML(anim.to_html5_video())
# HTML(anim.to_jshtml())  # 产生的文件太大，测试时可以打开
```
