```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1, 1, 0.001)  # np.arange(-1, 1.001, 0.001)，如果这样，下面计算y会出现负数
x = np.append(x, 1)
y1 = (1 - x**2) ** (1/2)
y2 = y1 * -1

plt.plot(x, y1)
plt.plot(x, y2)
# plt.axis("equal")  # 这样设置x y轴等比例，会使xlim与ylim设置失效
ax = plt.gca()       # 这样设置x y轴等比例
ax.set_aspect(1)     # 这样设置x y轴等比例
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.show()
```
