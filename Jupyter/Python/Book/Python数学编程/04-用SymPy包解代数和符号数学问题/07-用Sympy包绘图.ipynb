# 用Sympy包绘图

`SymPy` 包可以直接绘制方程的图形。使用 `Sympy` 绘图不需要提前准备数值，而是可以直接根据表达式（数学方程）绘图，十分方便。  
`Sympy` 包在内部也是使用 `matplotlib` 绘制图形。

```python
from sympy import Symbol, symbols, factor, expand
from sympy.plotting import plot
import sympy
```

```python
x = Symbol('x')
expr = 2*x + 3
plot(expr)
```

```python
plot((2*x + 3), (x, -5, 5))  # 默认x取值范围是[-10, 10]，这里限制为[-5, 5]
```

```python
plot(2*x + 3, (x, -5, 5), title='A Line', xlabel='x', ylabel='2x + 3')  # 设置title等属性
```

```python
p = plot(2*x + 3, (x, -5, 5), title='A Line', xlabel='x', ylabel='2x + 3', show=False)  # 不显示图像
p.save('../data/line.png')  # 保存图片
```

## 绘制表达式

```python
expr = input('Enter an expression: ')  # 2*x + 3*y - 6
expr = sympy.sympify(expr)
y = Symbol('y')
sympy.solve(expr, y)
```

```python
solution = sympy.solve(expr, y)  # [2 - 2*x/3]
expr_y = solution[0]

plot(expr_y)
```

## 绘制多个函数

```python
x = Symbol('x')
plot(2*x+3, 3*x+1)
```

```python
x = Symbol('x')
p = plot(2*x+3, 3*x+1, legend=True, show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p.show()
```
