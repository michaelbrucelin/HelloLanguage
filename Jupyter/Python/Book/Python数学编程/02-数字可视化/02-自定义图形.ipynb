# 自定义图形

```python
from pylab import *
```

## 1. 添加标题和标签

```python
nyc_temp_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]
nyc_temp_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]
nyc_temp_2012 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]
months = range(1, 13)
plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)
title('Average monthly temperature in NYC')  # 标题
xlabel('Month')                              # 横轴标签
ylabel('Temperature')                        # 纵轴标签
legend([2000, 2006, 2012])                   # 图例
show()
```

## 2. 自定义坐标轴

```python
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
plot(nyc_temp, marker='o')
axis()  # 返回值为4个数字的元组，前两个数字为x轴的范围，后两个数字为y轴的范围
```

```python
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
plot(nyc_temp, marker='o')
axis(ymin=0)  # 让y轴从0开始，axis([0, 10, 0, 20])一次性设置全部的值
show()
```

## 3. 使用pyplot绘图

`pylab` 模块可用于在交互式 `shell` 中绘图，例如 `IDLE shell`，然而当在大型程序中绘图时，`pyplot` 则更有效率。

```python
import matplotlib.pyplot as plt
```

```python
def create_graph():
    x_numbers = [1, 2, 3]
    y_numbers = [2, 4, 6]
    plt.plot(x_numbers, y_numbers)
    plt.show()

if __name__ == '__main__':
    create_graph()
```

## 4. 保存图形

```python
x = [1, 2, 3]
y = [2, 4, 6]
plot(x, y)  # 这里使用的是pylab下的plot()
savefig('../data/mygraph.png')
```
