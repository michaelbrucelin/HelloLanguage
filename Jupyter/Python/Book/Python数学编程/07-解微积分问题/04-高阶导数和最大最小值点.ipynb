# 高阶导数和最大最小值点

```python
import sympy as sp
```

```python
x = sp.Symbol('x')
f = x**5 - 30*x**3 + 50*x

sp.plotting.plot(f, (x, -5, 5))
```

```python
d1 = sp.Derivative(f, x).doit()     # 1阶导
d2 = sp.Derivative(f, x, 2).doit()  # 2阶导

p = sp.plotting.plot(f, d1, d2, (x, -5, 5), show=False)
p[0].line_color = 'b'
p[1].line_color = 'g'
p[2].line_color = 'r'
p.show()
```

```python
d1 = sp.Derivative(f, x).doit()  # 默认计算的就是1阶导数，1阶导数为0的点，就是驻点（包括局部极值点与全局极值点，或鞍点）
critical_points = sp.solve(d1)
critical_points                  # 结果为上图中的4个极值点
```

```python
A = critical_points[2]  # 四个点从左到右的顺序
B = critical_points[0]
C = critical_points[1]
D = critical_points[3]
```

```python
d2 = sp.Derivative(f, x, 2).doit()     # 计算2阶导数，2阶导 < 0，极大值点，2阶导 > 0，极小值点，2阶导 = 0，鞍点
print(f'A: {d2.subs({x:A}).evalf()}')  # 极大值
print(f'B: {d2.subs({x:B}).evalf()}')  # 极小值
print(f'C: {d2.subs({x:C}).evalf()}')  # 极大值
print(f'D: {d2.subs({x:D}).evalf()}')  # 极小值
```

```python
# 找出函数在某个区域的最大值和最小值，除了一阶导数为0的点外，还要把定义域边界的点考虑进去
x_min, x_max = -5, 5
print(f'x_min: {f.subs({x:x_min}).evalf()}')
print(f'x_max: {f.subs({x:x_max}).evalf()}')
print(f'A: {f.subs({x:A}).evalf()}')
print(f'B: {f.subs({x:B}).evalf()}')
print(f'C: {f.subs({x:C}).evalf()}')
print(f'D: {f.subs({x:D}).evalf()}')
```
