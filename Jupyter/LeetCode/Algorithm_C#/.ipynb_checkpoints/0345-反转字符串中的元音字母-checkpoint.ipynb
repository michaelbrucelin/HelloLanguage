# 345. 反转字符串中的元音字母

编写一个函数，以字符串作为输入，反转该字符串中的元音字母（a、e、i、o、u）。

## 示例 1：
> **输入：**"hello"  
> **输出：**"holle"

## 示例 2：
> **输入：**"leetcode"  
> **输出：**"leotcede"

## 提示：
- 元音字母不包含字母 "y" 。

[链接：https://leetcode-cn.com/problems/reverse-vowels-of-a-string/](https://leetcode-cn.com/problems/reverse-vowels-of-a-string/)

```
public class Solution {
    public static string ReverseVowels(string s) {
        char[] sarr = s.ToCharArray();
        HashSet<char> vowel = new HashSet<char>("aAeEiIoOuU");

        int start = 0, end = sarr.Length - 1;
        char swap;
        while (start < end)
        {
            if (vowel.Contains(sarr[start]))
            {
                while (start < end)
                {
                    if (vowel.Contains(sarr[end]))
                    {
                        swap = sarr[start];
                        sarr[start] = sarr[end];
                        sarr[end] = swap;

                        end--;
                        break;
                    }
                    end--;
                }
            }
            start++;
        }

        return new string(sarr);
    }
}
```

```
Console.WriteLine(Solution.ReverseVowels("hello"));
Console.WriteLine(Solution.ReverseVowels("leetcode"));
```
