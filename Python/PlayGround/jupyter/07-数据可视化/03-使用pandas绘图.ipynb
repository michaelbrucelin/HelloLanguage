```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 折线图

```python
# Series默认使用索引作为绘图的x轴，可以使用use_index=False来禁用这个功能
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()  # <==> s.plot.line()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
df.plot()  # <==> df.plot.line()
```

### 柱状图

```python
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

data.plot.bar(ax=axes[0], color='k', alpha=0.7)   # 垂直柱状图
data.plot.barh(ax=axes[1], color='k', alpha=0.7)  # 水平柱状图
```

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))

df.plot.bar()  # 在DataFrame的柱状图中，每一行的数据被分组在一起
```

```python
df.plot.barh(stacked=True, alpha=0.8)  # 堆积柱状图
```

```python
s = pd.Series(np.random.randint(0, 10, 100))
s.value_counts().sort_index().plot.bar()  # 使用value_counts()可以轻松的对Series值的频率进行可视化
```

### 直方图和密度图

- 直方图（histogram）是一种可以对值频率进行离散化显示的柱状图。数据点被拆分到离散的、间隔均匀的“箱”中，绘制的是各“箱”中数据点的数量。
- 密度图是一种与直方图相关的一种图表类型，它是通过计算“可能会产生观测数据的连续概率分布的估计”而产生的。一般的过程是将该分布近似为一组核（即诸如正态分布之类的较为简单的分布）。因此，密度图也被称作KDE（Kernel Density Estimate，核密度估计）图。

```python
tips = pd.read_csv('data/tips.csv')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])

tips['tip_pct'].plot.hist(bins=50)  # 直方图
```

```python
tips['tip_pct'].plot.density()
```
