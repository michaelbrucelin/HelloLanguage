```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.arange(10)
plt.plot(data)  # 这个操作后天默认创建figure和subplot
```

```python
fig = plt.figure()              # matplotlib所绘制的图像都位于Figure对象中
ax1 = fig.add_subplot(2, 2, 1)  # 不能直接在空的Figure中绘图，需要先创建一个或多个子图（subplot）才可以
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)  # 前两个参数表示Figure可以有多少个子图（这里是2*2），第三个参数表示这个子图在Figure中的位置
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax3 = fig.add_subplot(2, 2, 3)
ax2 = fig.add_subplot(2, 2, 2)
plt.plot([1.5, 3.5, -2, 1.6])   # matplotlib默认在最后一个使用的子图中绘制图像
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), 'k--')  # 'k--'表示绘制黑色虚线
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
plt.plot(np.random.randn(50).cumsum(), 'k--')
```
