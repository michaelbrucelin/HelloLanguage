```python
from sympy import Symbol, symbols, factor, expand
from sympy.plotting import plot
import sympy

# 使用Sympy绘图不需要提前准备数值，而是可以直接根据表达式（数学方程）绘图，十分方便
# Sympy在内部也是使用matplotlib绘制图形
```

```python
x = Symbol('x')
expr = 2*x + 3
plot(expr)
```

```python
plot((2*x + 3), (x, -5, 5))  # 默认x取值范围是[-10, 10]，这里限制为[-5, 5]
```
