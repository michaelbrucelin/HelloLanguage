```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 3x**2 + 4y**2 = 12
x = np.arange(-2, 2, 0.001)
x = np.append(x, 2)
y1 = (12-3*x**2)**(1/2)/2
y2 = y1 * -1

plt.plot(x, y1)
plt.plot(x, y2)
# plt.axis("equal")  # 这样设置x y轴等比例，会使xlim与ylim设置失效
ax = plt.gca()       # 这样设置x y轴等比例
ax.set_aspect(1)     # 这样设置x y轴等比例
plt.xlim(-2.4, 2.4)
plt.ylim(-2.4, 2.4)
plt.show()
```
