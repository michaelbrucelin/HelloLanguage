```python
from sympy import Symbol, symbols, factor, expand
import sympy
```

```python
x = Symbol('x')
expr = 2*x + 3
sympy.plot(expr)
```

```python
x = Symbol('x')
expr = x**2
sympy.plot(expr)
```

```python
x = Symbol('x')
expr1 = (1 - x**2)**(1/2)
expr2 = -1*((1 - x**2)**(1/2))
sympy.plot(expr1, expr2)
```

```python
x = Symbol('x')
y = Symbol('y')
expr = (1 - x**2 - y**2)**(1/2)
sympy.plotting.plot3d(expr)
```
