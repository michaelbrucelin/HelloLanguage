```python
from sympy import Symbol, symbols, factor, expand
from sympy.plotting import plot
import sympy

# 使用Sympy绘图不需要提前准备数值，而是可以直接根据表达式（数学方程）绘图，十分方便
# Sympy在内部也是使用matplotlib绘制图形
```

```python
x = Symbol('x')
expr = 2*x + 3
plot(expr)
```

```python
plot((2*x + 3), (x, -5, 5))  # 默认x取值范围是[-10, 10]，这里限制为[-5, 5]
```

```python
plot(2*x + 3, (x, -5, 5), title='A Line', xlabel='x', ylabel='2x + 3')  # 设置title等属性
```

```python
p = plot(2*x + 3, (x, -5, 5), title='A Line', xlabel='x', ylabel='2x + 3', show=False)  # 不显示图像
# p.save('line.png')  # 保存图片
```

```python
# 绘制表达式
x, y = symbols('x,y')
expr = 2*x + 3*y -6
expr = sympy.sympify(expr)
solution = sympy.solve(expr, y)  # [2 - 2*x/3]
solution[0]

plot(solution[0])
```

```python
# 绘制多个函数
x = Symbol('x')
plot(2*x+3, 3*x+1)
```

```python
x = Symbol('x')
p = plot(2*x+3, 3*x+1, legend=True, show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p.show()
```
