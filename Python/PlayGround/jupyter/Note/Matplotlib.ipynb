## 一、Matplotlib API 概观

我相信很多读者包括我，在一开始接触到matplotlib时一定会遇到plt.plot和ax.plot到底有什么区别？为什么有时候要用plt.show有时候不用？  
其实在matplotlib的API（Application Programming Interface）共分成两种型态，一种是我们常见到的plt.plot或是df.plot的格式，另一种则是ax.plot的格式。  
plt.plot：绘图的懒人包（包括 df.plot），会帮你直接在最近使用的图表上绘图，没有的话就会自动新建一个。  
ax.plot：手动建立的API，提供多项客制化的参数与图表控制。  
以我个人的经验，我会强烈建议先以ax.plot的完整API来学习，弄懂背后的完整原理，再来使用简单的df.plot懒人包。

## 二、Matplotlib的核心绘图观念



在绘图之前，要先搞清楚不同参数代表的意思。  
Figure：可以把他当作画布，一张画布上可以有很多的Axes。  
Axes：就是我们俗称的「子图」，每个Axes一次只能在一张画布上。  
这时就有人说，不对啊？我之前都用plt.plot来绘图，根本没有用到figure和axes。plt.plot其实是一个黑盒子，它帮你把基础的参数设定好，它就会在最近一张子图生成图表（没有则会创建一张）；但这不代表背后就没有figure和axes的运作。

### Add_subplot: 如何生成 “Figure” and “Axes”?

这里就来实际操作一次画布figure与子图axes的生成：

```python
# 将 figure 实例化到 fig 的变数上，并设定宽与高（此处为预设值）
fig = plt.figure(figsize=(6,4))

# 将 axe 新增到实例化后 2×2 的 figure 上，并指定第 1,2,3 个
ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)
```
