```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 使用python内建的模块实现一个简单的随机漫步
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
plt.plot(walk[:100])
```

```python
# 使用numpy改进上面的随机漫步
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws>0, 1, -1)
walk = steps.cumsum()
plt.plot(walk[:100])
```
