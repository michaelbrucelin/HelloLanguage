```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

pandas对象中的数据可以通过一些方式进行合并：
- pandas.merge可根据一个或多个键将不同DataFrame中的行连接起来，它实现的就是数据库的join操作。
- pandas.concat可以沿着一条轴将多个对象堆叠到一起。
- combine_first实例方法可以将重复数据拼接在一起，用一个对象中的值填充另一个对象中的缺失值。

### 数据库风格的join联接

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})

# inner join
pd.merge(df1, df2, on='key')  # 如果没有指定参数on，merge()会将重叠列的列名当做键。不过，最好明确指定指定
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})

# inner join
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
# left join
pd.merge(df1, df2, on='key', how='left')
```

```python
# right join
pd.merge(df1, df2, on='key', how='right')
```

```python
# full join
pd.merge(df1, df2, on='key', how='outer')
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})

pd.merge(left, right, on=['key1', 'key2'], how='outer')  # 根据多个键进行join
```

```python
pd.merge(left, right, on='key1')  # key2作为列名重复了，pandas自动补全了_x, _y作为尾缀
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))  # 使用suffixes选项设定重复列名的尾缀
```
