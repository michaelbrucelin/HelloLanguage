```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df
```

```python
melted = pd.melt(df, ['key'])  # 'key'为分组指标，其余列全部为值列
melted
```

```python
reshaped = melted.pivot('key', 'variable', 'value')  # 还原
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])  # 指定某些列为值列
```

```python
pd.melt(df, value_vars=['A', 'B', 'C'])  # 可以不指定分组指标
```
