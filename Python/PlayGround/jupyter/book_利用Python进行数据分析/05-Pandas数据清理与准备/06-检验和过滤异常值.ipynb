```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
data[2][np.abs(data[2])>3]  # 找出第三列中绝对值大于3的值
```

```python
data[(np.abs(data)>3).any(1)]  # 找出含有绝对值大于3的行
```

```python
data[np.abs(data)>3] = np.sign(data)*3  # sign()返回数值的正负值符号，将值限制在-3与3之间
data.describe()
```
