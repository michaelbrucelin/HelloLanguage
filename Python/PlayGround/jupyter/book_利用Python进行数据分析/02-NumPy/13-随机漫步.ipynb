```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 使用python内建的模块实现一个简单的随机漫步
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
plt.plot(walk[:100])
```

```python
# 使用numpy改进上面的随机漫步
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws>0, 1, -1)
walk = steps.cumsum()
plt.plot(walk[:100])
```

```python
walk.min()
```

```python
walk.max()
```

```python
# 找出第一次距离初始位置10步远（任意方向）的时间
(np.abs(walk)>=10).argmax()
# argmax()的效率并不高，因为它总是扫描整个数组，而在这个特定场景中，只要找到True就可以停止扫描了
```

```python
# 一次性模拟多次随机漫步
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws>0, 1, -1)
walks = steps.cumsum(1)
```

```python
walks.max()
```

```python
walks.min()
```

```python
# 计算穿越30或-30的最小穿越时间
hits30 = (np.abs(walks) >= 30).any(1)
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)
crossing_times.mean()  # 平均穿越时间
```
